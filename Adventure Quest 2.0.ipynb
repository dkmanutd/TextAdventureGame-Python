```python
"""
##############################################################################

Welcome to Adventure Quest - Treasure of Gringotts! In this game, you will
embark on a dangerous journey through the Woods of Muir, which if you
successfully survive, will lead you to discover the mythical treasure of
Gringotts! There will be three rounds: in the first, you will have to answer a
random question to check whether or not you get a bonus life; in the second, 
you will have to face three obstacles from across several multiverses; and
finally, you need to successfully walk through a minefield for this abundant 
treasure to be yours. God speed!

##############################################################################
"""

# Importing packages
import random
import time
import sys

# Declaring global variables
global player_weapon
player_weapon = ''
global obstacles
obstacles = ['Sugar Daddy', 'Sandman', 'Charmander', 'Firefly', 'Robot',
'Zombie', 'pretty girl', 'cute puppy']
```

```python
def delay_print(string):
    """
    Help on user-defined function delay_print:

    delay_print(...)
        delay_print(string)

        Prints the argument 'string' character by character.
    """

    for character in string:
        sys.stdout.write(character)
        sys.stdout.flush()
        time.sleep(0.02)
```

```python
def starting_path():
    """
    Help on user-defined function starting_path:

    starting_path()

        Takes the player through the first stage of the game, and depending on
        user input, either waits (and then loops back to the start of the
        function), proceeds to the next stage of the game, or jumps to the
        fail() function.
    """

    delay_print(f"""\nAs you stand at the edge of the woods, would you like \
to:\n1. stick around and admire the beauty of Mother Nature, or \n2. proceed \
to begin your attempt through the Woods of Muir, or \n3. flee in cowardice \
as many before you?\n""")

    response = input(prompt = '> ')
    response = response.lower()

    if '1' in response or 'stick' in response or 'stay' in response or 'stand' \
in response or 'chill' in response or 'wait' in response or 'admire' in \
response:
        delay_print(f"""\nIt really is beautiful here, isn't it {player_name}\
? It really makes you begin to wonder if it's worth risking your life for a \
mythical treasure..\n\n""")
        time.sleep(0.50)
        delay_print("<Press Enter when you are ready to reconsider>\n")
        input()
        starting_path()

    elif '2' in response or 'proceed' in response or 'go' in response or 'do' \
in response or 'play' in response:
        delay_print(f"""\nOnwards to the treasure of Gringotts, we go!\n\n""")
        time.sleep(0.50)
        delay_print("<Press Enter to continue>\n")
        input()

    elif '3' in response or 'flee' in response or 'run' in response or 'away' \
in response:
        fail()

    else:
        delay_print(f"""There was something wrong with your selection, \
{player_name}. Please try again.\n\n""")
        starting_path()
```

```python
def weapon_assign():
    """
    Help on user-defined function weapon_assign:

    weapon_assign()

        Simulates the roll of a die through the random.randint() function to
        generate a random number between 1 and 6, based on which the player is
        assigned a certain weapon. Function returns 'weapon'.
    """

    delay_print(f"""\nTo determine whether it is truly your destiny to \
discover the mythical treasure of Gringotts, a weapon will be assigned to \
you with which you must make your way through the Woods of Muir. Roll this \
magical die to be assigned your weapon..\n""")
    print("""
       ________
      /\\       \\
     /()\\   ()  \\
    /    \\_______\\
   \\    /()      /
    \\()/   ()   /
     \\/______()/

    """)
    time.sleep(0.50)
    delay_print("<Press Enter to roll die>\n")
    input()

    die_face = random.randint(1,6)

    if die_face == 1 or die_face == 2:
        weapon = 'Ice Sword'
        delay_print(f"""{player_name}, you have rolled a {die_face} on the \
magical die, and as such, you are being granted: {weapon}. Hopefully this \
weapon doesn't crumble to pieces.. Just "chill" but proceed with caution!\n""")
    elif die_face == 3 or die_face == 4:
        weapon = 'Unlimited Water Balloons'
        delay_print(f"""{player_name}, you have rolled a {die_face} on the \
magical die, and as such, you are being granted: {weapon}. Do not \
underestimate the effectiveness of H2O!\n""")
    else:
        weapon = 'Shotgun'
        delay_print(f"""{player_name}, you have rolled a {die_face} on the \
magical die, and as such, you are being granted: {weapon}. You will need to \
take careful aim with this weapon - happy hunting!\n""")

    return(weapon)
```

```python
def beginning():
    """
    Help on user-defined function beginning:

    beginning()

        Gives the user a chance to gain a bonus life by picking a question from
        'questions[]' using the random.choice() function, and then checks if the
        answer is correct to return a Boolean value of True or False. Function
        returns health_0.
    """
    
    questions = ['What part of the body produces insulin?', 'What type of \
animal is Bambi?', 'What is the name of the poker hand containing three \
of a kind and a pair?', 'Who was the Greek/Roman God of War?', 'Who is \
the greatest superhero of all time?', 'What is the capital of Turkey?', \
'Name the title of the 1818 novel by Mary Wollstonecraft Shelley.']

    question = random.choice(questions)
    delay_print(question + "\n")
    answer_0 = input("> ")
    answer_0 = answer_0.lower()
    health_0 = False

    if(question == questions[0]):
        if(answer_0 == 'pancreas'):
            health_0 = True
    if(question == questions[1]):
        if(answer_0 == 'deer'):
            health_0 = True
    if(question == questions[2]):
        if(answer_0 == 'full house'):
            health_0 = True
    if(question == questions[3]):
        if(answer_0 == 'ares'):
            health_0 = True
# I'm a huge Batman fan, so there's really no debating this next answer..
    if(question == questions[4]):
        if(answer_0 == 'batman'):
            health_0 = True
    if(question == questions[5]):
        if(answer_0 == 'ankara'):
            health_0 = True
    if(question == questions[6]):
        if(answer_0 == 'frankenstein'):
            health_0 = True

    if health_0 == True:
        delay_print(f"\n{player_name}, your answer is correct - you now have \
a bonus life in addition to your current life!\n\n")
    else:
        delay_print(f"\n{player_name}, unfortunately your response was \
incorrect so you are going to have to make your way through with the one life \
that you have..\n\n")

    return(health_0)
```
